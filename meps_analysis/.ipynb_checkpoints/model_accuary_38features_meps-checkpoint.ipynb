load some libraries

```python
import pandas
import numpy.linalg as lg
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import matplotlib as mpl
from sklearn.preprocessing import OneHotEncoder
from sklearn import linear_model
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import cross_val_predict
%matplotlib inline
```

OneHotEncoder function to convert the categorical features to numeric features
e.g. Male is 0, Female is 1
Input: X is pandas data frame

```python
def tonumeric(lst):
   lbls = {}
   for t in lst:
       if t not in lbls:
           lbls[t] = len(lbls.keys())
   return np.array([lbls[t] for t in lst])

def onehotencode_strings(X):
   cat_ft=[i for i in range(X.shape[1]) if 'str' in str(type(X.ix[0,i])) or 'unicode' in str(type(X.ix[0,i]))]
   if len(cat_ft): 
       for i in cat_ft:
           X.ix[:,i] = tonumeric(X.ix[:,i]) 
       X = OneHotEncoder(categorical_features=cat_ft).fit_transform(X)
   return X
```

Feature correlation matrix
Input: data is pandas data frame

Load csv data into pandas data frame

```python
data = pandas.read_csv("DATA PATH")
```

```python
#The data points from age below 10 contain too many missing values. Thus, exclude them from analysis
kept_index = data.AGE31X > 10 
```

```python
data = data[kept_index]
```

test_y = log(actual Y). Then, set -inf to zero

```python
Y = np.log(data.TOTEXP12)
Y = Y.values
```

```python
temp = Y[0]
for i in range(len(Y)):
    if Y[i] == temp:
        Y[i] = 0
```

Load top feature list in csv formate with header

```python
top_feature = pandas.read_csv("FEATURE PATH")
```

```python
top_feature.shape
```

```python
training_X = data.ix[:, top_feature.ix[:,0]]
training_X.shape
```

Convert categorical feature in X to numeric feature

```python
onehotencode_strings(training_X)
```

Ramdonly divide the dataset into training (75%) and testing(25%)

```python
X_train, X_test, y_train, y_test = train_test_split(training_X, Y)
```

Model performance in TRAINING dataset

```python
X_test.shape
```

```python
y_test.shape
```

```python
X = X_train.values
Y = y_train
```

5 fold cross-validation of the actual Y with predicted Y

```python
lr = linear_model.LinearRegression()
predicted = cross_val_predict(lr, X, Y, cv=5)
fig, ax = plt.subplots()
ax.scatter(Y, predicted)
ax.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'k--', lw=4)
ax.set_xlabel('Measured Log Transformed Expenditure')
ax.set_ylabel('Predicted Log Transformed Expenditure')
```

R squared values of linear fit

```python
lr.fit(X, Y)
lr.score(X, Y)
```

Pairwise feature correlation matrix, blue is low and red is high

```python
feature_correcoef(X)
```

```python
residutal = Y - predicted
plt.scatter(predicted, residutal,alpha=0.5)
plt.title('Residual Plot')
plt.xlabel('Predicted Y')
plt.ylabel('Residual')
```

Model performance in TEST data set

```python
X = X_test.values
Y = y_test
```

```python
lr = linear_model.LinearRegression()
predicted = cross_val_predict(lr, X, Y, cv=10)
fig, ax = plt.subplots()
ax.scatter(Y, predicted)
ax.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'k--', lw=4)
ax.set_xlabel('Measured Log Transformed Expenditure')
ax.set_ylabel('Predicted Log Transformed Expenditure')
```

```python
score = cross_val_score(lr, X, Y, cv=5)
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
```

Random Forest Modeling

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import cross_val_score
#have tried number of estimator from 5-500. 50 tested to give good results
estimator = RandomForestRegressor(random_state=0, n_estimators=50)
score = cross_val_score(estimator, X, Y, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
```
